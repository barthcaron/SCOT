```python
import sys
sys.path.insert(1, '../src/')
import utils as ut
import evals as evals
import scot2 as scot
import numpy as np
```

```python
# Loading the datasets (version pre-processed according to the original publication of the data)
X=np.load("../data/sci_atac_feat.npy") 
y=np.load("../data/sci_rna_feat.npy")
print("Dimensions of input datasets are:\n", "X (chromatin accessibility)= ", X.shape, "\n y (gene expression) = ", y.shape)

Xlabels=np.load("../data/sci_atac_labels.npy") 
ylabels=np.load("../data/sci_rna_labels.npy")

import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

pca=PCA(n_components=2)
Xpca=pca.fit_transform(X)
ypca=pca.fit_transform(y)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 4.5),
                               tight_layout=True, sharey=True)

ax1.scatter(Xpca[:,0], Xpca[:,1], c=Xlabels, cmap="rainbow")
ax2.scatter(ypca[:,0], ypca[:,1], c=ylabels, cmap="rainbow")
plt.show()
```

```python
unique, counts = np.unique(Xlabels, return_counts=True)

unique, counts = np.unique(ylabels, return_counts=True)
```

```python
sc=scot.SCOT(X, y)
X_new, y_new = sc.align(k=50, e=0.0005,  normalize=True)
Xpca=pca.fit_transform(X_new)
ypca=pca.fit_transform(y_new)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 4.5),
                               tight_layout=True, sharey=True)

ax1.scatter(Xpca[:,0], Xpca[:,1], c=Xlabels, cmap="rainbow")
ax2.scatter(ypca[:,0], ypca[:,1], c=ylabels, cmap="rainbow")
plt.show()


plt.scatter(ypca[:,0], ypca[:,1], c=ylabels, cmap="rainbow")
plt.scatter(Xpca[:,0], Xpca[:,1], c=Xlabels, cmap="rainbow")
plt.show()
#k=50, e=0.0005, score= , labelTransfer@5= 
```
